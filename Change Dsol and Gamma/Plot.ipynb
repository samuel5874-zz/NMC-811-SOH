```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_excel('0.3C 0.8C.xlsx')

# Extract data
cycle_number = df.iloc[:, 0]  # First column as x-axis
A = df.iloc[:, 1]  # Second column A
B = df.iloc[:, 2]  # Third column B

# Plot curves
plt.figure(figsize=(8, 6))
plt.plot(cycle_number, A, label='0.8C', marker='o')
plt.plot(cycle_number, B, label='0.3C', marker='s')

# Add title and labels
plt.xlabel('Cycle Number')
plt.ylabel('Remaining Capacity')
plt.title('Remaining Capacity vs. Cycle Number')
plt.legend()
plt.grid()

# Display plot
plt.show()
```
