```python
%pip install "pybamm[plot,cite]" -q

import pybamm
import matplotlib.pyplot as plt

# 定义电池模型
model = pybamm.lithium_ion.DFN(
    {
        "SEI": "solvent-diffusion limited",
        "calculate discharge energy": "true",  # for compatibility with older PyBaMM versions
    }
)

# 设定模型参数
param = pybamm.ParameterValues("OKane2022")
param.update({
    "SEI solvent diffusivity [m2.s-1]": 1.25e-20  # 新的值可以根据需求调整
})

var_pts = {
    "x_n": 5,  # 负极
    "x_s": 5,  # 隔膜
    "x_p": 5,  # 正极
    "r_n": 30,  # 负极颗粒
    "r_p": 30,  # 正极颗粒
}

# 设置最大循环次数
max_cycles = 300  # 你可以增加到 2000，但计算量会很大
cycle_numbers = [1] + list(range(50, max_cycles + 1, 50))  # 选择每50个循环计算一次
remaining_capacity_values = []  # 存储剩余容量

# 逐步增加循环次数进行模拟
for k in cycle_numbers:
    print(f"正在计算循环 {k} ...")  # 显示当前进度

    # 定义实验方案
    exp_k = pybamm.Experiment(
        [
            "Hold at 4.2 V until C/100 (5 minute period)",
            "Rest for 4 hours (5 minute period)",
            "Discharge at 0.1C until 2.5 V (5 minute period)",  # 初始容量检查
            "Charge at 0.3C until 4.2 V (5 minute period)",
            "Hold at 4.2 V until C/100 (5 minute period)",
        ]
        + [
            (
                "Discharge at 1C until 2.5 V",  # 老化循环
                "Charge at 0.3C until 4.2 V (5 minute period)",
                "Hold at 4.2 V until C/100 (5 minute period)",
            )
        ]
        * k
        + ["Discharge at 0.1C until 2.5 V (5 minute period)"],  # 最终容量检查
    )

    # 运行仿真
    sim_k = pybamm.Simulation(model, parameter_values=param, experiment=exp_k, var_pts=var_pts)
    sol_k = sim_k.solve()

    # 提取 LLI 损失（单位：A.h）并计算 Remaining Capacity (%)
    Q_LLI_final_k = sol_k["Total lithium lost [mol]"].entries[-1] * 96485.3 / 3600  # 转换单位
    remaining_capacity_k = (5 - Q_LLI_final_k) / 5 * 100  # 计算剩余容量百分比
    remaining_capacity_values.append(remaining_capacity_k)  # 存储数据

# 绘制 Remaining Capacity (%) vs. Cycle Number (k)
plt.figure(figsize=(8, 6))
plt.plot(cycle_numbers, remaining_capacity_values, label="Remaining Capacity (%)", color="blue", linewidth=2)
plt.xlabel("Number of cycles (k)")
plt.ylabel("Remaining Capacity (%)")
plt.title("Remaining Capacity vs. Cycle Number")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)  # 限制 y 轴范围在 0% - 100%
plt.show()

# 输出最后一个循环的剩余容量
print(f"当 k={cycle_numbers[-1]} 时，剩余容量为: {remaining_capacity_values[-1]:.2f}%")
```
